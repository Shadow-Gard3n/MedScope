## Drug List

```python
drug_list.info()
```

```python
print(f"Before: {len(drug_list)}")
drug_list = drug_list[(drug_list['role_cod'] == 'PS') | (drug_list['role_cod'] == 'SS')].copy()
print(f"After removing other than 'Primary Suspect' or 'Secondary Suspect' drugs: {len(drug_list)}")
```

```python
drug_list['drugname'].value_counts()
```

```python
drug_list_cleaned = drug_list.drop(columns = ['caseid','nda_num','lot_num','val_vbm','dose_vbm','exp_dt'])  # case id is still important when merging
drug_list_cleaned
```

```python
drug_list_cleaned.isnull().sum()
```

```python
drug_list_cleaned.drop(columns=['cum_dose_chr', 'cum_dose_unit','rechal','dose_freq'], inplace=True)   # too many null
drug_list_cleaned
```

```python
columns_to_rename = {
    'drug_seq': 'drug_sequence',
    'role_cod': 'drug_role',
    'prod_ai': 'active_chemical',
    'dechal': 'dechallenge',
    'dose_amt': 'dose_amount',
}
drug_list_cleaned = drug_list_cleaned.rename(columns=columns_to_rename)
drug_list_cleaned
```

```python
drug_list_cleaned['dose_unit'].value_counts()
```

```python
conversion_to_mg = {
    'G': 1000,    # g to mg
    'UG': 0.001,  # micro to mg
    'KG': 1000000, # kg to mg
    'MG': 1        
}

for unit, factor in conversion_to_mg.items():
    mask = drug_list_cleaned['dose_unit'] == unit     # when matched with unit in dictionary
    drug_list_cleaned.loc[mask, 'dose_amount'] *= factor
    drug_list_cleaned.loc[mask, 'dose_unit'] = 'MG'

drug_list_cleaned['dose_unit'].value_counts()
```

```python
drug_list_cleaned
```

```python
drug_list_cleaned = drug_list_cleaned[drug_list_cleaned['dose_unit'] == 'MG'].copy()   # Removing all other for analysis
drug_list_cleaned
```

```python
drug_list_cleaned.drop(columns = ['dose_unit'], inplace=True)  # now not needed
drug_list_cleaned
```

```python
drug_list_cleaned['dechallenge'].value_counts()
```

```python
drug_list_cleaned.isnull().sum()
```

```python
drug_list_cleaned.dropna(subset=['dose_amount'], inplace=True)
drug_list_cleaned['dechallenge'] = drug_list_cleaned['dechallenge'].fillna('U')
drug_list_cleaned['route'] = drug_list_cleaned['route'].fillna('Unknown')
drug_list_cleaned['dose_form'] = drug_list_cleaned['dose_form'].fillna('Unknown')

drug_list_cleaned.isnull().sum()
```

```python
dechallenge_map = {
    'U': 'Unknown',
    'Y': 'Yes',
    'N': 'No',
    'D': 'Not applicable'  # Does not apply
}
drug_list_cleaned['dechallenge'] = drug_list_cleaned['dechallenge'].map(dechallenge_map)
drug_list_cleaned
```

```python
drug_list_cleaned.info()
```

```python
drug_list_cleaned['primaryid'].value_counts()
```

## Report Details

```python
report_details.info()
```

```python
report_details.sort_values(by=['caseid', 'caseversion'], inplace=True)  # sort by caseid and version
report_details_cleaned = report_details.drop_duplicates(subset='caseid', keep='last').copy() # drop duplicates case
```

```python
report_details_cleaned['primaryid'].value_counts()
```

```python
report_details_cleaned.drop(columns=['caseid','caseversion','auth_num','lit_ref','to_mfr','mfr_num','mfr_sndr','event_dt','mfr_dt','init_fda_dt','rept_dt','i_f_code','e_sub','rept_cod'], inplace=True)
report_details_cleaned
```

```python
report_details_cleaned.info()
```

```python
report_details_cleaned.isnull().sum()
```

```python
report_details_cleaned['age_cod'].value_counts()
```

```python
import numpy as np
age_conversion_factors = {
    'YR':  1,
    'DEC': 10,    # decade
    'MON': 1 / 12,   # month
    'WK':  1 / 52.14,   # week
    'DY':  1 / 365.25,    # day
    'HR':  1 / 8766    # hour
}

report_details_cleaned['age_in_years'] = np.nan

for code, factor in age_conversion_factors.items():
    mask = report_details_cleaned['age_cod'] == code
    report_details_cleaned.loc[mask, 'age_in_years'] = report_details_cleaned.loc[mask, 'age'] * factor

report_details_cleaned[
    (report_details_cleaned['age_cod'].notna()) & (report_details_cleaned['age_cod'] != 'YR')
][['age', 'age_cod', 'age_in_years']].sample(10)
```

```python
report_details_cleaned.drop(columns=['age_cod','age'], inplace = True)
report_details_cleaned
```

```python
report_details_cleaned['wt_cod'].value_counts()
```

```python
lbs_to_kg_factor = 0.453592
lbs_mask = report_details_cleaned['wt_cod'] == 'LBS'
report_details_cleaned.loc[lbs_mask, 'wt'] = report_details.loc[lbs_mask, 'wt'] * lbs_to_kg_factor
report_details_cleaned.loc[lbs_mask, 'wt_cod'] = 'KG'
report_details_cleaned
```

```python
report_details_cleaned.drop(columns=['wt_cod'], inplace = True)
report_details_cleaned
```

```python
report_details_cleaned['occp_cod'].value_counts()
```

```python
hcp_codes = ['MD', 'PH', 'RN', 'OT']
report_details_cleaned['is_hcp'] = report_details_cleaned['occp_cod'].isin(hcp_codes)   # heatlh care professional
report_details_cleaned.drop(columns=['occp_cod'], inplace=True)
report_details_cleaned
```

```python
report_details_cleaned['sex'] = report_details_cleaned['sex'].fillna('UNK')
report_details_cleaned['sex'].value_counts()
```

```python
report_details_cleaned['reporter_country'].value_counts()
```

```python
report_details_cleaned['reporter_country'] = report_details_cleaned['reporter_country'].fillna('COUNTRY NOT SPECIFIED')
report_details_cleaned['occr_country'] = report_details_cleaned['reporter_country'].fillna('COUNTRY NOT SPECIFIED')
```

```python
report_details_cleaned.isnull().sum()
```

```python
report_details_cleaned.drop(columns=['wt'], inplace=True)  # too many nulls
report_details_cleaned
```

```python
median_age = report_details_cleaned['age_in_years'].median()
print(median_age)
report_details_cleaned['age_in_years'] = report_details_cleaned['age_in_years'].fillna(median_age)
```

```python
report_details_cleaned['age_grp'].value_counts()
```

```python
age_bins = [0, 1/12, 2, 12, 18, 65, 150]
age_labels = ['Neonate', 'Infant', 'Child', 'Adolescent', 'Adult', 'Elderly']

report_details_cleaned['age_grp'] = pd.cut(
    x=report_details_cleaned['age_in_years'], 
    bins=age_bins, 
    labels=age_labels, 
    right=False,
    include_lowest=True # To include age 0
)
report_details_cleaned
```

```python
report_details_cleaned.isnull().sum()
```

```python
report_details_cleaned[report_details_cleaned['age_grp'].isnull()]
```

```python
report_details_cleaned[report_details_cleaned['age_in_years']<0]
```

```python
unrealistic_age = (report_details_cleaned['age_in_years'] < 0) | (report_details_cleaned['age_in_years'] > 120)
report_details_cleaned.loc[unrealistic_age, 'age_in_years'] = median_age
report_details_cleaned.loc[unrealistic_age, 'age_grp'] = 'Adult'
```

```python
report_details_cleaned.isnull().sum()
```

```python
report_details_cleaned.info()
```
